## 실측데이터 11616개 , round == 1 데이터 개수 11616개,  round == 2 데이터 개수 11592개

```python
pred_0_1 = pred[(pred['model_id']==0)&(pred['round']==1)]
pred_0_2 = pred[(pred['model_id']==0)&(pred['round']==2)]
pred_1_1 = pred[(pred['model_id']==1)&(pred['round']==1)]
pred_1_2 = pred[(pred['model_id']==1)&(pred['round']==2)]
pred_2_1 = pred[(pred['model_id']==2)&(pred['round']==1)]
pred_2_2 = pred[(pred['model_id']==2)&(pred['round']==2)]
pred_3_1 = pred[(pred['model_id']==3)&(pred['round']==1)]
pred_3_2 = pred[(pred['model_id']==3)&(pred['round']==2)]
pred_4_1 = pred[(pred['model_id']==4)&(pred['round']==1)]
pred_4_2 = pred[(pred['model_id']==4)&(pred['round']==2)]
```

```python
# 'time' 변수를 기준으로 inner 조인
check_0_1 = pd.merge(pred_0_1, gens, on='time', how='inner')
check_0_2 = pd.merge(pred_0_2, gens, on='time', how='inner')
check_1_1 = pd.merge(pred_1_1, gens, on='time', how='inner')
check_1_2 = pd.merge(pred_1_2, gens, on='time', how='inner')
check_2_1 = pd.merge(pred_2_1, gens, on='time', how='inner')
check_2_2 = pd.merge(pred_2_2, gens, on='time', how='inner')
check_3_1 = pd.merge(pred_3_1, gens, on='time', how='inner')
check_3_2 = pd.merge(pred_3_2, gens, on='time', how='inner')
check_4_1 = pd.merge(pred_4_1, gens, on='time', how='inner')
check_4_2 = pd.merge(pred_4_2, gens, on='time', how='inner')
```

## round == 1 에는 있고 round == 2에는 없는 'time' 변수의 값

```python
# 'mae_check_0_1' 데이터프레임의 'time' 변수 값 세트로 변환
check_0_1_time_set = set(check_0_2['time'])

# 'gens' 데이터프레임에서 'time' 변수 값 가져오기
gens_time_set = set(gens['time'])

# 'gens' 데이터프레임에서 'time' 변수의 값 중 'mae_check_0_1' 데이터에 없는 값을 찾기
missing_times = [time for time in gens_time_set if time not in check_0_1_time_set]

# 결과 출력
print("Missing 'time' values in check_0_1 data:")
print(missing_times)
```

```python
check_0 = pd.concat([check_0_1, check_0_2], axis=0)
check_1 = pd.concat([check_1_1, check_1_2], axis=0)
check_2 = pd.concat([check_2_1, check_2_2], axis=0)
check_3 = pd.concat([check_3_1, check_3_2], axis=0)
check_4 = pd.concat([check_4_1, check_4_2], axis=0)
```

```python
check_0['예측오차율'] = check_0.apply(lambda row: 
                                            (abs((check_0.loc[(check_0['time'] == row['time']) & 
                                                                   (check_0['round'] == 1), 
                                                                  'pred_amount'].mean() - row['amount']) 
                                                 * (100 / 99))), axis=1)

check_1['예측오차율'] = check_1.apply(lambda row: 
                                            (abs((check_1.loc[(check_1['time'] == row['time']) & 
                                                                   (check_1['round'] == 1), 
                                                                  'pred_amount'].mean() - row['amount']) 
                                                 * (100 / 99))), axis=1)

check_2['예측오차율'] = check_2.apply(lambda row: 
                                            (abs((check_2.loc[(check_2['time'] == row['time']) & 
                                                                   (check_2['round'] == 1), 
                                                                  'pred_amount'].mean() - row['amount']) 
                                                 * (100 / 99))), axis=1)

check_3['예측오차율'] = check_3.apply(lambda row: 
                                            (abs((check_3.loc[(check_3['time'] == row['time']) & 
                                                                   (check_3['round'] == 1), 
                                                                  'pred_amount'].mean() - row['amount']) 
                                                 * (100 / 99))), axis=1)

check_4['예측오차율'] = check_4.apply(lambda row: 
                                            (abs((check_4.loc[(check_4['time'] == row['time']) & 
                                                                   (check_4['round'] == 1), 
                                                                  'pred_amount'].mean() - row['amount']) 
                                                 * (100 / 99))), axis=1)
```

```python
model_0 = check_0[check_0['round']==1][['time', '예측오차율']]
model_1 = check_1[check_1['round']==1][['time', '예측오차율']]
model_2 = check_2[check_2['round']==1][['time', '예측오차율']]
model_3 = check_3[check_3['round']==1][['time', '예측오차율']]
model_4 = check_4[check_4['round']==1][['time', '예측오차율']]
```

```python

# 'time' 열을 인덱스로 설정
pred.set_index('time', inplace=True)


# 시간대 별 'amount' 선 그래프 그리기
plt.figure(figsize=(12, 6))  # 그래프 크기 설정 (선택사항)
plt.plot(pred.index, pred.values, marker='o', linestyle='-')
plt.xlabel('시간')  # x축 레이블 설정
plt.ylabel('Amount')  # y축 레이블 설정
plt.title('시간대 별 Amount')  # 그래프 제목 설정
plt.grid(True)  # 그리드 표시 (선택사항)
plt.show()  # 그래프 출력
```

```python
# 'model_id'를 기준으로 데이터를 분리하고 각 'model_id'에 대한 그래프를 별도로 그리기
unique_model_ids = pred['model_id'].unique()

# 각 'model_id'별로 subplot을 만들어 그래프 그리기
fig, axs = plt.subplots(len(unique_model_ids), 1, figsize=(12, 6*len(unique_model_ids)), sharex=True)

# 각 'model_id'에 대해 그래프를 그립니다.
for i, model_id in enumerate(unique_model_ids):
    model_data = pred[pred['model_id'] == model_id]
    axs[i].plot(model_data.index, model_data['pred_amount'], marker='o', linestyle='-', label=f'Model ID {model_id}')
    axs[i].set_ylabel('Amount')  # y축 레이블 설정
    axs[i].legend()  # 범례 표시
    axs[i].grid(True)  # 그리드 표시 (선택사항)

plt.xlabel('시간')  # x축 레이블 설정
plt.suptitle('Model ID 별 시간대 별 Amount', y=0.92)  # 전체 그래프 제목 설정
plt.show()  # 그래프 출력
```
