```python
from typing import List

def calc_profit(actual_gens: List[float], forecast_gens: List[float]):
    CAPACITY = 99.0
    facility_utilization_rate = [actual / CAPACITY for actual in actual_gens]

    filter_facility_utilization_rate = [
        utilization >= 0.1 for utilization in facility_utilization_rate
    ]

    errors = [
        abs(forecast - actual) / CAPACITY * 100
        for forecast, actual in zip(forecast_gens, actual_gens)
    ]

    target_errors = [
        error
        for error, is_filtered in zip(errors, filter_facility_utilization_rate)
        if is_filtered
    ]
    target_actual_gens = [
        actual
        for actual, is_filtered in zip(
            actual_gens, filter_facility_utilization_rate
        )
        if is_filtered
    ]

    profits = [0] * len(target_actual_gens)

    for i, error in enumerate(target_errors):
        if error <= 6:
            profits[i] = target_actual_gens[i] * 4
        elif 6 < error <= 8:
            profits[i] = target_actual_gens[i] * 3

    return profits
```

```python
import statsmodels.api as sm

# 시간대별로 회귀분석을 수행하고 결과를 저장할 리스트를 생성합니다.
time_periods = merged_data['hour'].unique()
significant_periods = []

# 각 시간대별로 회귀분석을 수행합니다.
for time_period in time_periods:
    subset = merged_data[merged_data['hour'] == time_period]
    X = subset['cloud']
    y = subset['amount']
    X = sm.add_constant(X)
    
    model = sm.OLS(y, X).fit()
    
    # 회귀분석 결과 중 cloud의 p-value를 확인하여 유의수준(예를 들어, 0.05)보다 작으면 유의미한 영향이 있다고 판단합니다.
    if model.pvalues['cloud'] < 0.05:
        significant_periods.append(time_period)

significant_periods
```

## 10시모델

```python
import optuna
from optuna import Trial, visualization
from optuna.samplers import TPESampler
X = weather_weight_10.drop(columns=['round', 'pred_incentive', 'pred_amount', 'amount', '예측오차율'], axis=1)
y = weather_weight_10[['pred_amount']]
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, shuffle=True)#ㅌxgb용
import xgboost as xgb  

def objectivexgb(trial: Trial, X_train, y_train, X_val, y_val):
    param = {
        'n_estimators' : trial.suggest_int('n_estimators', 300, 7000,1),
        'max_depth' : trial.suggest_int('max_depth', 3, 20, 1),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2),
        'gamma': trial.suggest_float('gamma', 0.0, 0.1),
        'reg_alpha': trial.suggest_float('reg_alpha', 0.0, 0.1),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 0.1),
        'min_child_weight': trial.suggest_int('min_child_weight', 3, 15, 1),
        'random_state': 42,
        'early_stopping_rounds': 300,      
    }
    
    
    # 학습 모델 생성
    model = xgb.XGBRegressor(**param, n_jobs=-1)
    meta_model = model.fit(X_train, y_train, eval_set=[(X_val, y_val)],
                          verbose=50 )
    
    # 모델 성능 확인
    prediction = meta_model.predict(X_val)


    # 성능 평가
    stacking_rmse = np.sqrt(mean_squared_error(y_val, prediction))

    
    return stacking_rmse
```

```python
# 하이퍼 파라미터 튜닝

study = optuna.create_study(direction='minimize',sampler=TPESampler()) #sampler=TPESampler()
study.optimize(lambda trial : objectivexgb(trial, X_train, y_train, X_val, y_val), n_trials = 50)

print('Best trial : stacking_rmse {}, \nparams {}'.format(study.best_trial.value, study.best_trial.params))
```

```python
# 최적의 하이퍼 파라미터 저장 
params1 = study.best_trial.params
params1
```

```python
# 최종 모델 학습 
stack_meta_model = xgb.XGBRegressor(**params1)
stack_meta_model.fit(X, y)
```
