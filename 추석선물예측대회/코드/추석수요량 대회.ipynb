```python
import os
import pandas as pd
import random
import numpy as np
from sklearn.preprocessing import LabelEncoder


from autogluon.tabular import TabularDataset, TabularPredictor
import autogluon.core as ag

import warnings
warnings.filterwarnings("ignore")

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(42) 
```

```python
train=pd.read_csv("C:/Users/user/Desktop/데이콘/추석선물예측대회/open/train.csv")
test=pd.read_csv("C:/Users/user/Desktop/데이콘/추석선물예측대회/open/test.csv")
```

```python
train = train.drop(['ID'], axis = 1)
test = test.drop('ID', axis = 1)
```

```python
train_data = TabularDataset(train)
test_data = TabularDataset(test)

label = '수요량'
eval_metric = 'root_mean_squared_error'
time_limit = 3600*7
```

```python
predictor = TabularPredictor(
    label=label, eval_metric=eval_metric
).fit(train_data,  presets='best_quality', time_limit=time_limit, 
      num_stack_levels=5)
```

```python
model_to_use = predictor.get_model_best()
model_pred = predictor.predict(test_data, model=model_to_use)
```

```python
sample=pd.read_csv("C:/Users/user/Desktop/데이콘/추석선물예측대회/open/sample_submission.csv")
```

```python
sample['수요량'] = model_pred
```
